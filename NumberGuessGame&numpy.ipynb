```python
# Number guessing game. Try it. Guess a number between 1 to 100. Max 4 chances, if you get it, then you might have good intuition.
```

```python
import random
b = random.randint(1,100)
Guess_count = 0
while True:
    try:
        n = int(input())
    except ValueError:
        print("Error, please try again!") 
        continue
    if n>b:
        print("Your guess is on higher side, please re-try")
        Guess_count+=1
    if n<b:
        print("Your guess is on lower side, please re-try")
        Guess_count+=1
    if Guess_count == 4:
        print("You have reached maximum tries")
        print(f"Correct answer was {b}")
        break
    if n==b:
        print("Hurrah you guessed it right")
        break
```

```python
## vertical flip the picture
b = camaro[::-1, :, :]
plt.imshow(b)
plt.show()
```

```python
io.imsave("Vertical_Flip.jpg",b)
```

```python
## horizontal flip the picture
c = camaro[:, : :-1, :]
plt.imshow(camaro)
plt.show()
```

```python
# Red color
red = np.zeros(camaro.shape, dtype = "int")
red[: ,:,0] = camaro[:,:,0]
plt.imshow(red)
plt.show()
```

```python
green = np.zeros(camaro.shape, dtype = "int")
green[: ,:,1] = camaro[:,:,1]
plt.imshow(green)
plt.show()
```

```python
blue = np.zeros(camaro.shape, dtype = "int")
blue[: ,:,2] = camaro[:,:,2]
plt.imshow(blue)
plt.show()
```

```python
rainbow = np.vstack((red,green,blue))
plt.imshow(rainbow)
plt.show()
```
